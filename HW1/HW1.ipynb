```python
import random
import re

def readfile(file):
    with open (file, encoding='utf-8') as f:
        lines = f.readlines()
        tasks = []
        for line in lines:
            line2_0 = line.strip('\n')
            task = line2_0.strip(' ')
            tasks.append(task)
        return tasks

def givequest(number):
    if number == 1:
        response = readfile("Lingterms.txt")
    elif number == 2:
        response = readfile("Professions.txt")
    else:
        response = " "
    return response

def distribute(data):
    if type(data) == list:
        feedback = random.choice(data)
    else:
        feedback = "Такой темы у нас нет!"
    return feedback

def hide(line):
    symbollist = []
    for letter in line:
        symbol = letter.replace(letter, "_ ")
        symbollist.append(symbol)
    task = "".join(symbollist)
    return task

def tellattempts(number):
    if number > 4:
        print("У Вас осталось", number, "попыток")
    elif number > 1:
        print("У Вас осталось", number, "попытки")
    elif number == 1:
        print("У Вас осталось", number, "попытка")
    else:
        print("Вы проиграли!")

def reveal(word, letterlist):
    symbollist = []
    for i, symbol in enumerate(word):
        if symbol in letterlist:
            newsym = symbol
        else:
            newsym = "_ "
        symbollist.append(newsym)    
    task = "". join(symbollist)
    return task

def gameplay():
    hanglist = ['  |', ' ()', '/()\\', ' /\\', '/  \\']
    details = []
    hacked = []
    trash = []
    answer = int(input("Выберите тему: лингвистические термины(введите 1) или профессии(введите 2)? Ваш ответ: "))
    taskbank = givequest(answer)
    question = distribute(taskbank)
    task = hide(question)
    if question == "Такой темы у нас нет!":
        print(question)
    else:
        print(task)
        while len(hanglist) > 0:
                tellattempts(len(hanglist))
                guess = input("Какая буква есть?")
                if len(guess) != 1:
                    print("Вводите строго по одной букве!")
                elif re.search('[a-zA-Z0-9А-ЯЁ]', guess):
                    print("Вводите только строчные буквы русского языка")
                elif guess in hacked or guess in trash:
                    print("Это Вы уже вводили! Не повторяйтесь.")
                elif guess in question:
                    hacked.append(guess)
                    new_task = reveal(question, hacked)
                    print(new_task)
                    if not re.search("_", new_task):
                        print("Вы выиграли!")
                        break
                else:
                    trash.append(guess)
                    details.append(hanglist[0])
                    for detail in details:
                        print(detail)
                    hanglist = hanglist[1::]
                    tellattempts(len(hanglist))
        else:
            print("Вы проиграли!")
            print(question)
            
gameplay()
```
